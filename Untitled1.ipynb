Statsmodels ile modelleme

```python
import statsmodels.api as sm
X = df[["TV"]]
X = sm.add_constant(X)
```

```python
y = df["sales"]
```

```python
lm = sm.OLS(y,X)
```

```python
model = lm.fit()
```

```python
model.summary()
```

```python
model.params
```

```python
model.summary().tables[1]
```

```python
model.conf_int()
```

```python
model.f_pvalue
```

```python
print("f_pvalue: ", "%.4f" % model.f_pvalue)
```

```python
print("fvalue: ", "%.2f" % model.fvalue)
```

```python
model.mse_model
```

```python
model.rsquared_adj
```

```python
model.fittedvalues[0:5]
# modelin tahmin değerleri
```

```python
y[0:5]
# gerçek değerler
```

```python
model.params[1]
```

```python
print("Sales = " + str("%.2f" % model.params[0]) + " + TV" + " * " + str("%.2f" % model.params[1]))
# Modelin denklemi 
```
